```python
from qiskit import *
from qiskit_textbook.tools import random_state, array_to_latex
from qiskit.visualization import plot_bloch_multivector
from qiskit.extensions import Initialize
```

```python
def entangle(circuit, q1, q2):
    circuit.h(q1)
    circuit.cx(q1, q2)
    
def alice(circuit, q1, q2):
    circuit.cx(q1, q2)
    circuit.h(q1)
    
def bob(circuit, qubit, cr1, cr2):
    circuit.x(qubit).c_if(cr1, 1)
    circuit.z(qubit).c_if(cr2, 1)
```

```python
def teleport(psi, init_gate):
    #setup initial circuit; q1 and q2 are given to Alice and Bob respectively by a third party post-entanglement. Alice's initial state is q0
    quantum_register = QuantumRegister(3)
    classical_register_one = ClassicalRegister(1)
    classical_register_two = ClassicalRegister(1)
    circuit = QuantumCircuit(quantum_register, classical_register_one, classical_register_two)
    circuit.append(init_gate, [0])
    
    #entangle qubits 1 and 2; these are the two qubits Alice and Bob receive
    entangle(circuit, 1, 2)
    circuit.barrier()
    
    #Apply a CNOT gate and Hadamard gate
    alice(circuit, 0, 1)
    circuit.barrier()
    
    #Measure q0 and q1
    for i in range(2):
        circuit.measure(i, i)
    circuit.barrier()
    
    #Now apply the right number of X and Z gates to retrieve psi
    bob(circuit, 2, classical_register_two, classical_register_one)
    
    return circuit
```

```python
#testing
psi = random_state(1)

init_gate = Initialize(psi)
init_gate.label = "init"

array_to_latex(psi, pretext="|\\psi\\rangle =")
plot_bloch_multivector(psi)
```

```python
circuit = teleport(psi, init_gate)
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(circuit, backend=simulator).result()
new_vectors = result.get_statevector()
plot_bloch_multivector(new_vectors)
```
