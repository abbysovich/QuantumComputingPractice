```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
```

```python
def create_oracle(secret_no):
    # We want to make a change to the last qubit if both the secret number and the qubit are in the state 1 at the same place value.
    n = len(secret_no)
    oracle = QuantumCircuit(n + 1)
    for i, digit in enumerate(secret_no):
        if digit == "1":
            oracle.cx(i, n)
    return oracle
```

```python
secret_number = '1010011100101'
n = len(secret_number)

qc = QuantumCircuit(n + 1, n)
qc.x(n)
for qubit in range(n + 1):
    qc.h(qubit)
qc.barrier()
qc.draw(output='mpl')
```

```python
qc += create_oracle(secret_number)
qc.barrier()
qc.draw(output='mpl')
```

```python
#Hadamarding everything will result in qubits 0 to n-1 being in state |s> with 100% chance. Here, s is the secret number.
for qubit in range(n):
    qc.h(qubit)
qc.barrier()
qc.draw(output='mpl')
```

```python
#All that is left is to measure the first n qubits
for qubit in range(n):
    qc.measure(qubit, qubit)
qc.draw(output='mpl')
```

```python
plot_histogram(execute(qc, Aer.get_backend('qasm_simulator')).result().get_counts())
```
